Now let's have a look at the non numerical data

```python
df['Destination'].value_counts()
df['HomePlanet'].value_counts()
```

```python
dest_to_num = {
    'TRAPPIST-1e': 1,
    '55 Cancri e': 2,
    'PSO J318.5-22': 3
}

homep_to_num = {
    'Earth': 1,
    'Europa': 2,
    'Mars': 3
}
df['Destination'] = df['Destination'].replace(dest_to_num)
df['HomePlanet'] = df['HoemPlanet'].replace(homep_to_num)

df['Destination'].value_counts()
```

```python
df['CryoSleep'] = df['CryoSleep'].replace({True: 1, False: 0})
df['VIP'] = df['VIP'].replace({True: 1, False: 0})
df['Transported'] = df['Transported'].replace({True: 1, False: 0})
df['VIP'].value_counts()
```

```python
corr_mtx = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_mtx, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix', fontsize=16)
plt.show()
```
